# Comienzo del ejercicio.

Muestro los resultados en cada celda para la revisión y sé que no es necesario.

Agrupo por categoría las fechas y sentimientos promedio.

```python
play_store_fechas = play_store.groupby("Category")["Last Updated"].mean().reset_index()
play_store_fechas
```

Uno los dataframes, conservando los datos de ambos. Puede que haya apps con nombres repetidos, pero como pueden tener datos diferentes, las conservo de todas maneras.

```python
play_store_sentimientos = pd.merge(play_store, play_store_user, how = "inner")
play_store_sentimientos
```

```python
play_store_sentimientos = play_store_sentimientos.groupby("Category")[["Sentiment_Polarity", "Sentiment_Subjectivity"]].mean().reset_index()
play_store_sentimientos
```

```python
play_store_sliced = pd.merge(play_store_sentimientos, play_store_fechas, how = "outer")
play_store_sliced
```

```python
play_store_sliced["Last Updated"] = (datetime.now() - play_store_sliced["Last Updated"]).dt.days
play_store_sliced
```

La correlación entre la antiguedad promedio de una categoría con el promedio de la polaridad de sus reviews y con el promedio de la subjetividad es:

```python
play_store_sliced.corr(numeric_only = True)
```

La correlación puede cambiar dependiendo el día y hora que se ejecute el notebook, ya que toma el datetime.now() para los cálculos de los días desde la última actualización.
