Muestro los resultados en cada celda para la revisión y sé que no es necesario.

Reemplazo "Varies with device" con NaN porque no me sirve a la hora de calcular los pesos de los géneros de las aplicaciones. Quito las k (kB) de los tamaños para tenerlos en números, y al final se los agregaré de nuevo. Para el caso de los M (MB), pasaré el tamaño a kB.

```python
play_store.loc[play_store["Size"].str.contains("k"), "Size"] = play_store.loc[play_store["Size"].str.contains("k"), "Size"].str.replace("k", "")
play_store.loc[play_store["Size"].str.contains("M"), "Size"] = play_store.loc[play_store["Size"].str.contains("M"), "Size"].str.replace("M", "").astype(float) * 1024
play_store["Size"] = play_store["Size"].replace("Varies with device", np.nan)
play_store["Size"] = play_store["Size"].astype(float)
```

Esto es para ver si se mantiene el peso de la última fila (58) que tiene el tamaño en kB.

```python
play_store.head(59)
```

Separo los géneros para obtener el peso porque hay más de uno en algunas filas.

```python
play_store["Genres"] = play_store["Genres"].str.split(";")
play_store = play_store.explode("Genres")
play_store
```

```python
play_store_generos = play_store.groupby("Genres")["Size"].sum().reset_index()
play_store_generos
```

Le agrego la unidad a los pesos totales de cada género y quedan representados en kB:

```python
play_store_generos["Size"] = play_store_generos["Size"].astype(str) + "k"
play_store_generos
```
