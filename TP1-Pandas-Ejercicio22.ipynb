Muestro los resultados en cada celda para la revisión y sé que no es necesario.

Agrupo sin tomar en cuenta las filas con NaN que no tienen datos con valores.

```python
play_store_sentimientos = play_store_user.groupby(["App", "Sentiment"]).size().unstack().reset_index()
play_store_sentimientos
```

```python
play_store_sentimientos.loc[play_store_sentimientos.isna().any(axis = "columns")]
```

Como tengo valores NaN, significa que hay 0 opiniones de ese tipo.

```python
play_store_sentimientos = play_store_sentimientos.fillna(0)
play_store_sentimientos.head(14)
```

```python
play_store_sentimientos["Total"] = play_store_sentimientos["Negative"] + play_store_sentimientos["Neutral"] + play_store_sentimientos["Positive"]
play_store_sentimientos["Positive+Negative"] = play_store_sentimientos["Negative"] + play_store_sentimientos["Positive"]
play_store_sentimientos["Neutral %"] = (play_store_sentimientos["Neutral"] * 100) / play_store_sentimientos["Total"]
play_store_sentimientos["Positive+Negative %"] = ((play_store_sentimientos["Positive"] + play_store_sentimientos["Negative"] ) * 100) / play_store_sentimientos["Total"]
play_store_sentimientos
```

Las 10 aplicaciones que generaron opiniones más polarizadas son:

```python
play_store_sentimientos.nlargest(10, ["Positive+Negative %", "Positive+Negative"])
```
